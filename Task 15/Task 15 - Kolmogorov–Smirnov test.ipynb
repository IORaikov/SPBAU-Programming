# Тест Колмогорова-Смирнова
Возьмём экспериментальные данные в файле data.csv и проверим, что они являются нормальным распределением.
Данные были получены в ходе работы 1.01 лабораторного практикума Политеха. Ход работы был следующим:

* Первый студент замечал на таймере 7 секунд, командуя вслух "Старт" и "Стоп" при начале и конце отсчета.
* Второй студент секундомером измерял этот промежуток времени, базируясь на командах первого.
* Третий записывал результаты измерений. 

Данные должны иметь вид нормального распределения со средним в районе 7 секунд (+- статическая погрешность), а отклонения от среднего демонстрируют роль человеческого фактора при проведении измерений.

```python
import pandas
from scipy import stats
dataset = pandas.read_csv("data.csv")['experiments']

print(f'Mean={dataset.mean()}, STD={dataset.std()}')
dataset.plot.kde()
stats.kstest(dataset,'norm',(dataset.mean(),dataset.std()))
```

Итак, график плотности распределения выглядит *нормально*, тест показывает низкую статистику и высокое p - то есть распределение близко к гауссовому. Это радует, потому что ту лабораторную нам зачли.
